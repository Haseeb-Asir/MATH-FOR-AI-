```python
import numpy as np 
import matplotlib.pyplot as plt 

data = np.random.uniform(0,1,1000)


plt.hist(data,bins = 10, density= True)
plt.title("Hist of uniform dist")
plt.show()
```

```python
from scipy.stats import lognorm

data = lognorm.rvs(s = 1 , scale = np.exp(0),size = 1000)

plt.hist(data,bins = 30, density= True, alpha =0.5)
plt.show()
```

```python
from scipy.stats import pareto

data = pareto.rvs(b=2,size = 1000)

plt.hist(data,bins = 30, density=True, alpha =0.9)
plt.xscale("log")
plt.yscale("log")
plt.title("Pareto distribution")
plt.show()
```

```python
from scipy.stats import bernoulli

# Generate Bernoulli data (p=0.3)
data = bernoulli.rvs(p=0.3, size=1000)

# Plot
plt.hist(data, bins=2, density=True)
plt.xticks([0, 1], ["Failure", "Success"])
plt.title("Bernoulli Distribution (p=0.3)")
plt.show()
```

```python
from scipy.stats import binom

# Generate Binomial data (n=10 trials, p=0.5)
data = binom.rvs(n=10, p=0.5, size=1000)

# Plot
plt.hist(data, bins=9, density=True, alpha=0.9)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
import pandas as pd

# Example: Right-skewed data
data = pd.Series(np.random.lognormal(mean=0, sigma=1, size=1000))

# Log transform
log_data = np.log(data)

# Box-Cox transform (requires positive data)
boxcox = PowerTransformer(method="box-cox")
boxcox_data = boxcox.fit_transform(data.values.reshape(-1, 1))

# Yeo-Johnson (handles any data)
yeojohnson = PowerTransformer(method="yeo-johnson")
yj_data = yeojohnson.fit_transform(data.values.reshape(-1, 1))

# Compare plots
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].hist(log_data, bins=30); axes[0].set_title("Log Transform")
axes[1].hist(boxcox_data, bins=30); axes[1].set_title("Box-Cox Transform")
axes[2].hist(yj_data, bins=30); axes[2].set_title("Yeo-Johnson Transform")
plt.show()
```

```python
# CLT Demo: Non-normal population (exponential)
population = np.random.exponential(scale=1, size=10000)

# Take 1000 samples of size n=30 and compute their means
sample_means = [np.mean(np.random.choice(population, 30)) for _ in range(1000)]

# Plot
plt.hist(population, bins=30, alpha=0.5, label="Population (Exponential)")
plt.hist(sample_means, bins=30, alpha=0.5, label="Sample Means (Normal)")
plt.legend()
plt.title("Central Limit Theorem in Action")
plt.show()
```
